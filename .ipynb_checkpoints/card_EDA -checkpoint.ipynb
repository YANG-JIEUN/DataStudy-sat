```python
import numpy as np
estimators = [(np.mean, 'Mean'), ## 평균
              (np.median, 'Median'), ## 중앙값
              (np.max, 'Max'), ## 최대값
              (np.min, 'Min') ## 최소값
             ]
sns.set_theme(style="whitegrid") ## figure, axes 스타일
fig, ax = plt.subplots(2, 2)
fig.set_size_inches((10,10))
fig.set_facecolor('white')
for i, e in enumerate(estimators):
    estimator = e[0]
    title = e[1]
    row = i//2
    col = i%2
    cur_ax = ax[row, col]
    sns.barplot(x="month", y="tco_btc_u_am", data=card_dt_man, hue='tco_btc_nm', estimator=estimator,
                ax=cur_ax, ## 현재 axes에 그려짐
                errwidth=0 
               )
    cur_ax.set_title(title)
```

### 정규성 검정 

#### 가설설정 
H_0 : 데이터 분포가 정규성을 따른다. (귀무가설) 
H_1 : 데이터 분포는 정규성을 따르지 않는다. (대립가설) 

** 귀무가설을 기각하고 대립가설이 채택되는 경우(p<0.01 or 0.05) 해당 데이터set 은 정규 분포를 따르지 않는 것이다. **




```python
rand1 = np.random.normal(loc=1, scale=2, size =1000)
rand2 = np.random.uniform(low=0, high = 2, size = 1000)
rand3 = np.random.f(dfnum=10, dfden = 10, size=1000)

plt.figure(figsize = (20,5))
sns.kdeplot(data=rand1 , color = 'red', shade =True)
sns.kdeplot(data=rand2 , color = 'blue', shade = True)
sns.kdeplot(data=rand3 , color = 'green', shade = True)
plt.show()


# 정규분포, 유니폼 분포, f 분포에서의 랜덤 난수를 생성 
```

```python
#정규성 검정 - 왜 정규성 검정을 하는가? 해당 데이터 셋의 신뢰성 검증을 위하여 분석에 신뢰성이 없다면 사용 가능한 인사이트 혹은 분석의 결과라고 할 수 없음. 
#표본이 많은 수가 아니기 때문에 qq-plot을 사용하여 검정 
import scipy.stats as stats

fig = plt.figure(figsize = (20,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

stats.probplot(rand2 , dist=stats.norm, plot = ax1)

mu = rand2.mean()
variance = rand2.var()
sigma = variance ** 0.5
x = np.linspace(mu - 3*sigma, mu + 3*sigma , 100)
ax2.plot(x, stats.norm.pdf(x,mu,sigma), color = "blue",label = "theoretical")

sns.distplot(ax=ax2, a= rand2, bins= 100 , color = 'red', label = 'observed')
ax2.legend()

plt.show()
```

```python
#정규성 분포에 대한 검정이 없었고 모든 변수가 자료형이 아니기때문에 상관관계를 spearman으로 확인해보기. 

df = card_data.corr(method='spearman')

# 그림 사이즈 지정
fig, ax = plt.subplots( figsize=(7,7) )

# 삼각형 마스크를 만든다(위 쪽 삼각형에 True, 아래 삼각형에 False)
mask = np.zeros_like(df, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# 히트맵을 그린다
sns.heatmap(df, 
            cmap = 'RdYlBu_r', 
            annot = True,   # 실제 값을 표시한다
            mask=mask,      # 표시하지 않을 마스크 부분을 지정한다
            linewidths=.5,  # 경계면 실선으로 구분하기
            cbar_kws={"shrink": .5},# 컬러바 크기 절반으로 줄이기
            vmin = -1,vmax = 1   # 컬러바 범위 -1 ~ 1
           )  
plt.show()

```
